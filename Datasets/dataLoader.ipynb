```python
from datasets import load_dataset
from transformers import AutoTokenizer
from huggingface_hub import notebook_login

notebook_login()
```

```python
import random
def get_wikitext2_train_loader(
    model_name,
    nsamples=128,
    seq_len=2048,
    seed=42
):
    dataset = load_dataset("wikitext", "wikitext-2-raw-v1", split="train")

    tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=False)

    text = "\n\n".join(dataset["text"])

    enc = tokenizer(text, return_tensors="pt")

    input_ids = enc.input_ids[0]

    random.seed(seed)
    trainloader = []

    for _ in range(nsamples):
        i = random.randint(0, input_ids.size(0) - seq_len - 1)
        inp = input_ids[i : i + seq_len].unsqueeze(0)
        tar = inp.clone()

        # Ignore loss for all tokens except last
        tar[:, :-1] = -100

        trainloader.append((inp, tar))

    return trainloader
```

```python
def get_loader(model_name, nsamples=128, seq_len=2048, seed=42, dataset='wikitext2'):
  if('wikitext2' in dataset):
    return get_wikitext2_train_loader(model_name, nsamples, seq_len, seed)
```
